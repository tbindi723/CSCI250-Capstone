```python
import numpy as np
import RPi.GPIO as GPIO
import adcUtil as adc
import time
import pigpio

# create reference to pigpio daemon
pi = pigpio.pi(port = 8887)

# setup GPIO as input for the gate pin
GPIO.setmode(GPIO.BCM)
switchPin = 26
buzPin1 = 12

buttonYellow = 24
buttonGreen = 18
buttonBlue = 22
buttonRed = 19

#make sure that the GPIO is set up 
GPIO.setup(buttonYellow, GPIO.IN, pull_up_down = GPIO.PUD_DOWN)
print(GPIO.input(buttonYellow))
GPIO.setup(buttonGreen, GPIO.IN, pull_up_down = GPIO.PUD_DOWN)
print(GPIO.input(buttonGreen))
GPIO.setup(buttonBlue, GPIO.IN, pull_up_down = GPIO.PUD_DOWN)
print(GPIO.input(buttonBlue))
GPIO.setup(buttonRed, GPIO.IN, pull_up_down = GPIO.PUD_DOWN)
print(GPIO.input(buttonRed))

# assign a variable for the pin number
ledPinYellow = 13
ledPinGreen = 21
ledPinBlue = 16
ledPinRed = 20

# set the GPIO pin to output mode
GPIO.setup(ledPinYellow, GPIO.OUT)
GPIO.setup(ledPinGreen, GPIO.OUT)
GPIO.setup(ledPinBlue, GPIO.OUT)
GPIO.setup(ledPinRed, GPIO.OUT)

#set up wrong and correct sounds
cdata = np.load("correctSong.npy")
correctFrequency = cdata[:,0]
correctDuration = cdata[:,1]

wdata = np.load("wrongSong.npy")
wrongFrequency = wdata[:,0]
wrongDuration = wdata[:,1]
```

```python
Vin =  3.3 # Volt
R2  = 10.0 # kOhm
R1 = 0

wL = False #if you lose it becomes true

Vou = adc.readADC(channel=1,device=0) # flex sensor voltage
if(Vou != 0):
    R1 = R2 * (Vin/Vou - 1)                # flex sensor resistance
    Rflex = 10 + R1; 
    
    
while True:
    #Get flex sensors resistor 
    Vou = adc.readADC(channel=1,device=1) # flex sensor voltage
    if(Vou != 0):
        R1 = R2 * (Vin/Vou - 1)                # flex sensor resistance 
    #if it is bent
    if R1 < Rflex:
        counter = 0  #only want to do this acertain amonent of times
        GPIO.setup(switchPin, GPIO.IN, pull_up_down = GPIO.PUD_UP)
        if (GPIO.input(switchPin) == 0): # if mag is connected to reed switch then send to hard Level
            print("mag")
            while (counter < 5) and (wL == False):
                wL = hardLevel(ledPinYellow, ledPinGreen, ledPinBlue, ledPinRed, counter, buttonYellow, buttonGreen, buttonBlue, buttonRed)
                if(wL == True):
                    wrongCorrectSong(buzPin1, wrongFrequency*2, wrongDuration) #buzzer plays wrong noise and breaks from loop
                    break
                else:
                    wrongCorrectSong(buzPin1, correctFrequency*2, correctDuration) #buzzer plays correct noise and continues in loop
                    counter +=1
            break
        else:# if mag is not connected to reed switch then send to easy level
            while (counter < 5) and (wL == False):  
                wL = easyLevel(ledPinYellow, ledPinGreen, ledPinBlue, ledPinRed, counter, buttonYellow, buttonGreen, buttonBlue, buttonRed)
                if(wL == True):
                    print("wrong")
                    wrongCorrectSong(buzPin1, wrongFrequency*2, wrongDuration) #buzzer plays wrong noise and breaks from loop
                    break
                    
                else:
                    print("correct")
                    wrongCorrectSong(buzPin1, correctFrequency*2, correctDuration) #buzzer plays correct noise and continues in loop
                    counter +=1
            break
    
```
